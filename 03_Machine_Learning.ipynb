````python
df.rename(columns = {'genre.2.1' : 'genre'}, inplace = True)
````

````python
df['genre'].value_counts()
````

````python
df = df.drop(columns=['id', 'title'])
````

````python
import ast 

def first_country(val):
    # works for strings that look like "['US', 'CA']" or "[]"
    if isinstance(val, str) and val.strip().startswith('['):
        try:
            lst = ast.literal_eval(val)
            return lst[0] if lst else None   
        except (ValueError, SyntaxError):
            return None                      
    return val                                

df['production_countries'] = df['production_countries'].map(first_country)
````

````python
df.head()
````

````python
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(include='object').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna('Unknown')
````

````python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
````

````python
low_card = [c for c in cat_cols if c != 'genre' and df[c].nunique() < 30]
for col in low_card:
    df[col] = LabelEncoder().fit_transform(df[col])
````

````python
df.head()
````

````python
X = df[['runtime', 'production_countries', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score', 'Revenue(In Millions)']]
y = df['genre']
X.head()
````

````python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.01, random_state=42
)
X_train.shape
````

````python
model = DecisionTreeClassifier(
    criterion="gini",             
)
model.fit(X_train, y_train)
````

````python
model.score(X_test,y_test)
````
