```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('Data_Clean_Titles.csv')
df2 = pd.read_csv('Data_Clean_Credits.csv')
```

```python
# FACT TABLE: Storing all the numeric values
fact_table = df1[['id', 'runtime', 'imdb_score', 'imdb_votes', 'tmdb_popularity',
                   'tmdb_score', 'Revenue(In Millions)']].reset_index(drop = True)

fact_table.to_csv('Fact_Table.csv', index = False)
fact_table.head()
```

```python
import ast
#Dimension Table 1 -> Genre Table
#Creating Genre Table, We are splitting every genre of a particular movie across its rows
# Convert the genre string to actual list

df1['genres'] = df1['genres'].apply(ast.literal_eval)
genre_table = df1[['id', 'genres']].explode('genres').rename(columns={'genres': 'genre'})

genre_table.reset_index(drop=True, inplace=True)

genre_table.to_csv('Genre_Table.csv', index=False)
genre_table.head()
```

```python
#Dimension Table 2 -> Title Table
title_dim = df1[['id', 'title', 'type', 'description', 'release_year']].reset_index(drop = True)

title_dim.to_csv('Title_Table.csv', index = False)
title_dim.head()
```

```python
df2.head()
```

```python
#Dimension Table 3 -> Character Name
character_dim = df2[['id', 'name', 'character', 'role']].reset_index(drop = True)

character_dim.to_csv('Character_Table.csv', index = False)
character_dim.head()
```

```python
fact_table.to_sql('fact_table', con=engine, if_exists='replace', index=False)
title_dim.to_sql('title_table', con=engine, if_exists='replace', index=False)
genre_table.to_sql('genre_table', con=engine, if_exists='replace', index=False)
character_dim.to_sql('character_table', con=engine, if_exists='replace', index=False)
```
